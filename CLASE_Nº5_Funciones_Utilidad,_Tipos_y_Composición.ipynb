Las funciones, dentro del campo de las matematicas, son herramientas esenciales que nos permiten modelar relaciones entre variables y comprender el comportamiento de los sistemas. Mas precisamente son reglas que asignan a cada valor de entrada **(dominio)** un único valor de salida **(codominio)**, creando así una correspondencia entre conjuntos. **1. Utilidad de las Funciones:**

Las funciones tienen varias utilidades y usos en diversos campos, incluyendo:

* **Ciencia e Ingeniería:** Modelar fenómenos físicos, químicos, eléctricos, entre otros.
* **Economía:** Representar relaciones entre variables económicas como precios, demanda y oferta.
* **Informática:** Desarrollar algoritmos, procesar datos y crear software.
* **Matemáticas:** Estudiar propiedades de números, figuras geométricas y conceptos abstractos.

Su utilidad radica en su capacidad para:

* **Simplificar problemas complejos:** Reducir situaciones con múltiples variables a relaciones entre dos variables.
* **Analizar comportamientos:** Estudiar la evolución de un sistema o fenómeno en función de una variable.
* **Resolver problemas:** Encontrar soluciones a problemas planteados en términos de funciones.
* **Comunicar ideas:** Expresar relaciones matemáticas de forma concisa y precisa.

**2. Tipos de Funciones:**

Existen varios tipos de funciones clasificadas según sus características:

* **Función lineal: f(x) = mx + b:** donde m y b son constantes. Representa una línea recta en el plano cartesiano.
* **Función cuadrática: f(x) = ax^2 + bx + c:** donde a, b y c son constantes. Representa una parábola en el plano cartesiano.
* **Función exponencial: f(x) = a^x:** donde a es una constante positiva. Representa una curva que crece o decrece exponencialmente.
* **Función logarítmica: f(x) = log_a(x):** donde a es una constante positiva y a ≠ 1. Representa la inversa de la función exponencial.
* **Función trigonométrica: f(x) = sin(x), cos(x), tan(x):** etc. Representan relaciones periódicas con el ángulo en radianes.
* **Función polinomial: f(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_1 x + a_0, donde a_n ≠ 0:** Representa una curva con múltiples cambios de dirección.

**3. Composición de Funciones:**

La composición de funciones es una operación matemática que combina dos o más funciones para crear una nueva función. Se representa con el símbolo "o".

**Notación:**

La composición de **f(x)** y **g(x)**, denotada como **f(g(x))** o **(f ◦ g)(x)**, se define como la función que aplica primero **g(x)** a cada valor de x y luego aplica **f(x)** al resultado.

**Ejemplo:**

Si consideramos f(x) = x^2 y g(x) = 2x + 1. La composición f(g(x)) = (f ◦ g)(x) se calcula como:

(f ◦ g)(x) = f(g(x)) = f(2x + 1) = (2x + 1)^2 = 4x^2 + 4x + 1

**Propiedades de la composición de funciones:**

* La composición de funciones es asociativa: (f ◦ g ◦ h)(x) = (f ◦ (g ◦ h))(x) = (f(g(h(x)))).
* La composición de funciones es conmutativa solo en algunos casos específicos.
