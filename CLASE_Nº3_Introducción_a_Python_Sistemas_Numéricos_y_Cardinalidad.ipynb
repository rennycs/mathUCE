**2.-Sistemas numéricos** El lenguaje de los números:

Los sistemas numéricos son la base para representar y manipular cantidades expresadas por ejemplo en numeros. A lo largo de la historia, las civilizaciones han desarrollado diversos sistemas numéricos, cada uno con sus propias características y aplicaciones.

**SISTEMAS NUMERICOS MAS UTILIZADOS**

**Sistema decimal:** El sistema numérico más utilizado en la actualidad, basado en 10 dígitos (0 a 9).

**Sistema binario:** El lenguaje fundamental de las computadoras, que utiliza solo dos dígitos (0 y 1).

**Sistemas numéricos no posicionales:** Sistemas como el romano o el babilónico, donde el valor de un dígito depende de su posición.

**Aritmética de números enteros y reales:** Operaciones básicas como suma, resta, multiplicación y división con números enteros y reales.

**¿Cómo podemos y para que podemos usar los sistemas numéricos?**

**Representar cantidades:**Expresar valores numéricos de forma precisa y concisa.

**Realizar cálculos:** Efectuar operaciones matemáticas con facilidad y eficiencia.

**Comprencion de la tecnología:** Decodificar el funcionamiento de las computadoras y dispositivos digitales.

**3.-Cardinalidad** Explorando el mundo de los conjuntos:

La cardinalidad es el estudio de la cantidad de elementos que componen un conjunto. Esta nos permite comprender la magnitud de conjuntos finitos e infinitos, y analizar las relaciones que existen entre ellos.

**CONCEPTOS CLAVE DE LA CARDINALIDAD**

**Conjuntos finitos:** Conjuntos con un número determinado de elementos.

**Conjuntos infinitos:** Conjuntos con un número incontable de elementos.

**Cardinalidad de conjuntos:** Medida de la cantidad de elementos de un conjunto.

**Operaciones con conjuntos:** Unión, intersección, diferencia y complemento de conjuntos.

**Principio de conteo:** Técnicas para determinar el número de elementos posibles en diferentes situaciones.

**USOS DE LA CARDINALIDAD**

**Organizar y clasificar información:** Agrupar elementos en función de sus características comunes.

**Resolver problemas de conteo:** Calcular el número de posibilidades en diferentes escenarios.

**Comprender conceptos matemáticos abstractos:**
Desarrollar un pensamiento lógico y riguroso.

**CODIGO PARA CONVERTIR UN NUMERO A BINARIO**


```python
def convertir_a_binario(numero):
  """
  Convierte un número entero a su representación binaria.

  Args:
    numero: El número entero a convertir.

  Returns:
    La representación binaria del número como cadena de caracteres.
  """
  return bin(numero)

numero_decimal = 10
numero_binario = convertir_a_binario(numero_decimal)
print(f"El número {numero_decimal} en binario es: {numero_binario}")
```
