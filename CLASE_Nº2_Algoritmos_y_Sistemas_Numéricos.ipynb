**DESCRIPCIONES BASES**

**Algoritmos**

Un algoritmo es o se trata de una serie de pasos ordenados y finitos que se utilizan para resolver un problema o realizar una tarea. Los algoritmos son la base de la programación y se utilizan en una amplia gama de aplicaciones, desde simples cálculos matemáticos hasta complejos sistemas de inteligencia artificial.

**Características de un algoritmo:**

**Precisión:** Cada paso debe estar definido de forma clara y sin ambigüedades.

**Finito:** El algoritmo debe terminar en un número finito de pasos.

**Eficiencia:** El algoritmo debe utilizar los recursos de manera eficiente, como tiempo y memoria.

**Generalidad:** El algoritmo debe ser aplicable a una variedad de problemas similares.


**Ejemplos de algoritmos:**

**Ordenar una lista de números:** Existen varios algoritmos para ordenar listas, como el algoritmo de burbuja, el algoritmo de inserción y el algoritmo de quicksort.

**Encontrar el camino más corto entre dos puntos: **Algoritmos como Dijkstra o A* se utilizan para encontrar el camino más corto en grafos.

**Buscar un elemento en una lista:** Algoritmos como la búsqueda lineal o la búsqueda binaria se utilizan para encontrar elementos en listas.

**Sistemas Numéricos**

Un sistema numérico es un conjunto de símbolos y reglas que se utilizan para representar números. Los sistemas numéricos más comunes son el sistema decimal **(base 10)**, el sistema binario **(base 2)** y el sistema hexadecimal **(base 16).**

**Ejemplos de sistemas numéricos:**

**Sistema decimal:** Utiliza los dígitos del 0 al 9 para representar números.

**Sistema binario:** Utiliza los dígitos 0 y 1 para representar números. Es el sistema utilizado por las computadoras.

**Sistema hexadecimal:** Utiliza los dígitos del 0 al 9 y las letras A a F para representar números. Se utiliza a menudo en programación y para representar colores en diseño web.

**Inteligencia artificial:** Los algoritmos se utilizan para crear sistemas de inteligencia artificial, como reconocimiento de voz y procesamiento de lenguaje natural.

```python
def bubble_sort(lista):
  n = len(lista)
  for i in range(n-1):
    for j in range(n-i-1):
      if lista[j] > lista[j+1]:
        lista[j], lista[j+1] = lista[j+1], lista[j]
  return lista

# Ejemplo de uso
lista_desordenada = [5, 1, 4, 2, 8]
lista_ordenada = bubble_sort(lista_desordenada.copy())

print("Lista desordenada:", lista_desordenada)
print("Lista ordenada:", lista_ordenada)
```
